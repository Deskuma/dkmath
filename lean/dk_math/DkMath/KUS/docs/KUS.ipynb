```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1.00, 20, 200)  # x >= 1 の範囲で定義
H = np.log(x) / x
plt.plot(x, H)
plt.axhline(0, color='red', linestyle='--')  # H(x) = 0 の線
plt.title('Graph of H(x) = log(x)/x')
plt.xlabel('x')
plt.ylabel('H(x)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Original (x, y) plane: contours of F(x,y) = x^y - y^x
x = np.linspace(1.0, 5.0, 800)
y = np.linspace(1.0, 5.0, 800)
X, Y = np.meshgrid(x, y)
F = np.power(X, Y) - np.power(Y, X)

fig, ax = plt.subplots(figsize=(8, 6))
cs = ax.contour(X, Y, F, levels=[0, 1])
ax.clabel(cs, inline=True, fontsize=9, fmt={0: "Gap = 0", 1: "Gap = 1"})
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title(r"Contours of $x^y - y^x$ for Gap = 0 and Gap = 1")
ax.grid(True, alpha=0.3)

# Known/sample points mentioned in the conversation
pts = np.array([
    [2.0, 1.0],
    [2.23924700014400258, 1.362806420000100],
    [3.0, 2.0],
])
ax.scatter(pts[:, 0], pts[:, 1], marker="o")
for px, py in pts:
    ax.annotate(f"({px:.5g}, {py:.5g})", (px, py), xytext=(5, 5), textcoords="offset points", fontsize=8)

xy_path = "KUS-gap_contours_xy.png"
fig.tight_layout()
fig.savefig(xy_path, dpi=300)
plt.close(fig)

# 2) Transformed (U, V) plane:
# U = y log x = log(x^y), V = x log y = log(y^x)
# Gap = 0  -> U = V
# Gap = 1  -> e^U - e^V = 1 -> U = log(1 + e^V)
V = np.linspace(-2.0, 6.5, 1000)
U_gap0 = V
U_gap1 = np.log1p(np.exp(V))

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(V, U_gap0, label="Gap = 0 : U = V")
ax.plot(V, U_gap1, label=r"Gap = 1 : $U=\log(1+e^V)$")
ax.set_xlabel("V = x log(y)")
ax.set_ylabel("U = y log(x)")
ax.set_title(r"Normalized $(U,V)$ plane for $x^y$ and $y^x$")
ax.grid(True, alpha=0.3)
ax.legend()

uv_path = "KUS-gap_contours_uv.png"
fig.tight_layout()
fig.savefig(uv_path, dpi=300)
plt.close(fig)

print(f"Saved: {xy_path}")
print(f"Saved: {uv_path}")
```
